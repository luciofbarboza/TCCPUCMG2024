## Model Training and Optimization

```python
# Exclusão dos dados realizados para não influenciar a predição com dados indevidos
X = df.drop('TML_LAB', axis=1)
Y = df['TML_LAB']
```

```python
X.tail()
```

```python
Y = pd.DataFrame(Y)
Y.tail()
```

```python
# Código para testar vários modelos de regressão usando o lazypredict

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Definir o modelo LazyRegressor
reg = LazyRegressor(verbose=0, ignore_warnings=True, custom_metric=None)

# Treinar e avaliar os modelos
models, predictions = reg.fit(X_train, X_test, y_train, y_test)

# Imprimir os resultados
print(models)
```

```python
# Código para avaliar o modelo SVR considerando as métricas r2, mae, mape e rmse

from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error

# Treinando o modelo SVR
svr_model = SVR(kernel='linear')
svr_model.fit(X_train, y_train)

# Fazendo predições com os dados teste
y_pred = svr_model.predict(X_test)

# Avaliando o modelo
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print("Avaliação do modelo SVR:")
print(f"R-squared: {r2}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Absolute Percentage Error (MAPE): {mape}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
# Gerar gráfico comparando y_test com o y_pred

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Valores Reais (y_test)')
plt.ylabel('Valores Preditos (y_pred)')
plt.title('Comparação entre Valores Reais e Preditos')
plt.grid(True)
plt.show()
```

```python
# Gerar gráfico de linha comparando y_test com o y_pred

plt.figure(figsize=(12, 6))
plt.plot(y_test.values, label='Valores Reais')
plt.plot(y_pred, label='Valores Preditos')
plt.xlabel('Índice')
plt.ylabel('TML_LAB')
plt.title('Comparação entre Valores Reais e Preditos')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Etapa de cross validation para verificar o desempenho do modelo SVR

from sklearn.model_selection import cross_val_score, KFold

# Número de folds
n_splits = 5

# Criando um objeto KFold
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Criar modelo SVR
svr_model = SVR(kernel='linear')

# Performar cross-validation e calcular metricas
adjusted_r2_scores = []
r2_scores = []
mae_scores = []
mape_scores = []
rmse_scores = []

for train_index, test_index in kf.split(X):
    X_train_fold, X_test_fold = X.iloc[train_index], X.iloc[test_index]
    y_train_fold, y_test_fold = Y.iloc[train_index], Y.iloc[test_index]

    svr_model.fit(X_train_fold, y_train_fold)
    y_pred_fold = svr_model.predict(X_test_fold)

    r2 = r2_score(y_test_fold, y_pred_fold)
    mae = mean_absolute_error(y_test_fold, y_pred_fold)
    mape = mean_absolute_percentage_error(y_test_fold, y_pred_fold)
    rmse = mean_squared_error(y_test_fold, y_pred_fold, squared=False)

    n = len(y_test_fold)
    p = X_train_fold.shape[1]
    adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

    adjusted_r2_scores.append(adjusted_r2)
    r2_scores.append(r2)
    mae_scores.append(mae)
    mape_scores.append(mape)
    rmse_scores.append(rmse)

# Imprimir os scores médios para todos os folds
print("Resultados do Cross-Validation (SVR):")
print(f"Média de Adjusted R-squared: {np.mean(adjusted_r2_scores):.4f}")
print(f"Média de R-squared: {np.mean(r2_scores):.4f}")
print(f"Média de MAE: {np.mean(mae_scores):.4f}")
print(f"Média de MAPE: {np.mean(mape_scores):.4f}")
print(f"Média de RMSE: {np.mean(rmse_scores):.4f}")
```

```python
# Ajuste de hiperparâmetros do modelo SVR

from sklearn.model_selection import GridSearchCV

# Definição do parametro grid para o modelo SVR
param_grid = {
    'kernel': ['linear'],  # Definição do Kernel
    'C': [0.1],  # Parametro de regularização
    'epsilon': [0.1]  # Epsilon in the epsilon-SVR model
}

svr_model = SVR()

# Criar objeto GridSearchCV
grid_search = GridSearchCV(estimator=svr_model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', verbose=2)

# Ajuste do grid search aos dados de treinamento
grid_search.fit(X_train, y_train)

print("Best parameters found:", grid_search.best_params_)

# Obtendo o melhor estimador
best_svr_model = grid_search.best_estimator_

# Avaliando o melhor modelo no conjunto de teste
y_pred = best_svr_model.predict(X_test)

# Calcular metricas para o melhor modelo
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print("Avaliação do modelo SVR (melhor modelo):")
print(f"R-squared: {r2}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Absolute Percentage Error (MAPE): {mape}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
