## Treinamento do modelo e otimização

```python
#Exclusão dos dados realizados do dataframe para não influenciar a predição com dados indevidos
X = df.drop('TML_LAB', axis=1)
Y = df['TML_LAB']
```

```python
X.tail()
```

```python
Y = pd.DataFrame(Y)
Y.tail()
```

```python
# Código para testar vários modelos de regressão usando o lazypredict

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Definir o modelo LazyRegressor
reg = LazyRegressor(verbose=0, ignore_warnings=True, custom_metric=None)

# Treinar e avaliar os modelos
models, predictions = reg.fit(X_train, X_test, y_train, y_test)

# Imprimir os resultados
print(models)
```

```python
# código para avaliar o modelo LinearRegression

from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
# Treinar o modelo LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = model.predict(X_test)

# Calcular as métricas
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Imprimir as métricas
print(f"R²: {r2}")
print(f"MAE: {mae}")
print(f"MAPE: {mape}")
print(f"RMSE: {rmse}")
```

```python
# Gráfico comparando y_test com o y_pred

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Valores Reais (y_test)')
plt.ylabel('Valores Preditos (y_pred)')
plt.title('Comparação entre Valores Reais e Preditos')
plt.grid(True)
plt.show()
```

```python
# Gráfico de linha comparando y_test com o y_pred

plt.figure(figsize=(12, 6))
plt.plot(y_test.values, label='Valores Reais')
plt.plot(y_pred, label='Valores Preditos')
plt.xlabel('Índice')
plt.ylabel('TML_LAB')
plt.title('Comparação entre Valores Reais e Preditos')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Cross validation para verificar o desempenho do modelo

from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
from sklearn.linear_model import LinearRegression

# Definir o modelo LinearRegression
model = LinearRegression()

# Definir as métricas para a validação cruzada
scoring = {
    'r2': 'r2',
    'mae': 'neg_mean_absolute_error',
    'mape': 'neg_mean_absolute_percentage_error',
    'rmse': 'neg_root_mean_squared_error',
    'adjusted_r2': 'r2'  # We'll calculate adjusted R-squared later
}

# Realizar a validação cruzada com 5 folds
cv_results = cross_validate(model, X, Y, cv=5, scoring=scoring)

# Calcular o Adjusted R-squared para cada fold
n = X.shape[0]  # Número de observações
k = X.shape[1]  # Número de variáveis independentes
adjusted_r2_scores = []
for r2_score_fold in cv_results['test_r2']:
    adjusted_r2 = 1 - (1 - r2_score_fold) * (n - 1) / (n - k - 1)
    adjusted_r2_scores.append(adjusted_r2)

# Imprimir os resultados da validação cruzada
print("Resultados da Validação Cruzada:")
print("Adjusted R²:", np.mean(adjusted_r2_scores))
print("R²:", np.mean(cv_results['test_r2']))
print("MAE:", -np.mean(cv_results['test_mae']))
print("MAPE:", -np.mean(cv_results['test_mape']))
print("RMSE:", -np.mean(cv_results['test_rmse']))
```

```python
# Ajuste de hiperparâmetros do modelo LinearRegression

from sklearn.model_selection import GridSearchCV

# Definir o modelo LinearRegression
model = LinearRegression()

# Definir a grade de hiperparâmetros a serem testados
param_grid = {
    'fit_intercept': [True, False],
    'positive': [True, False],
}

# Criar o objeto GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')

# Ajustar o modelo com a grade de hiperparâmetros
grid_search.fit(X_train, y_train)

# Imprimir os melhores hiperparâmetros encontrados
print("Melhores hiperparâmetros:", grid_search.best_params_)

# Avaliar o modelo com os melhores hiperparâmetros no conjunto de teste
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Calcular as métricas
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Imprimir as métricas
print(f"R²: {r2}")
print(f"MAE: {mae}")
print(f"MAPE: {mape}")
print(f"RMSE: {rmse}")
```

```python
# Exibindo novo gráfico prev x real após ajuste de hiperparametros
plt.figure(figsize=(12, 6))
plt.plot(y_test.values, label='Valores Reais')
plt.plot(y_pred, label='Valores Preditos')
plt.xlabel('Índice')
plt.ylabel('TML_LAB')
plt.title('Comparação entre Valores Reais e Preditos')
plt.legend()
plt.grid(True)
plt.show()
```
